```
import pandas as pd
import numpy as np

print("Iniciando el proceso de reemplazo de valores específicos en un dataset de 20 registros...\n")

# 1. Crear un dataset con 8 columnas que contenga información al azar con 20 registros
#    y que algunos campos contengan valores incorrectos o inconsistentes.
print("1. Creando un dataset de ejemplo con 20 registros y valores erróneos/inconsistentes:\n")

num_records = 20 # Número de registros solicitados

# Definimos los datos con algunas inserciones de valores incorrectos/inconsistentes
data = {
    'ID_Orden': np.arange(1001, 1001 + num_records),
    'Producto': np.random.choice(['Laptop', 'Teclado', 'Mouse', 'Monitor', 'Webcam', 'Microfono'], num_records),
    'Precio_Unitario': np.random.choice([599.99, 120.50, 25.00, -50.00, 750.00, 'NO_APLICA', 30.75, 1000.00], num_records),
    'Cantidad': np.random.choice([1, 2, 3, 0, 99], num_records), # 99 como valor erróneo
    'Estado_Envio': np.random.choice(['Enviado', 'Pendiente', 'Entregado', 'Cancelado', 'DESCONOCIDO', 'ERROR_LOGISTICA'], num_records),
    'Region': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste', 'NA'], num_records),
    'Calificacion_Servicio': np.random.choice([1, 2, 3, 4, 5, 'MALO'], num_records), # 'MALO' como valor inconsistente
    'Comentarios_Cliente': np.random.choice(['Ok', 'Problema', 'N/A', 'SIN_INFO', 'Excelente', 'Vacío'], num_records)
}

# Creamos el DataFrame
df = pd.DataFrame(data)

# Introducir más valores erróneos/inconsistentes intencionalmente en el dataset de 20 registros
df.loc[[2, 7, 15], 'Precio_Unitario'] = 'NO_DISP'
df.loc[[5, 10], 'Cantidad'] = -1
df.loc[[1, 12, 18], 'Estado_Envio'] = 'desconocido'
df.loc[[3, 9, 14], 'Region'] = 'No Aplica'
df.loc[[6, 11, 16], 'Calificacion_Servicio'] = 'ERROR'
df.loc[[4, 8, 13, 17], 'Comentarios_Cliente'] = 'n/a'


print("--- Dataset Original con Valores Incorrectos/Inconsistentes (20 Registros) ---\n")
print(df)

print("\n--- Conteo de Valores Específicos Antes del Reemplazo (Ejemplos) ---\n")
print("Valores en 'Precio_Unitario':\n", df['Precio_Unitario'].value_counts(dropna=False))
print("\nValores en 'Estado_Envio':\n", df['Estado_Envio'].value_counts(dropna=False))
print("\nValores en 'Cantidad':\n", df['Cantidad'].value_counts(dropna=False))
print("\nValores en 'Calificacion_Servicio':\n", df['Calificacion_Servicio'].value_counts(dropna=False))


# 2. Para el dataset que formulaste, escribe el codigo python que permita
#    reemplazar los valores erróneos por valores correctos usando replace() en las columnas que creaste.
print("\n2. Reemplazando valores erróneos/inconsistentes usando df.replace()...\n")

# Creamos una copia del DataFrame para no modificar el original directamente
df_cleaned = df.copy()

# --- Limpieza de la columna 'Precio_Unitario' ---
# Reemplazar strings no numéricos por NaN, luego convertir a numérico.
# Los precios negativos también se considerarán errores.
df_cleaned['Precio_Unitario'] = df_cleaned['Precio_Unitario'].replace(['NO_APLICA', 'NO_DISP', 'ERROR'], np.nan)
df_cleaned['Precio_Unitario'] = pd.to_numeric(df_cleaned['Precio_Unitario'], errors='coerce') # 'coerce' convierte errores a NaN
df_cleaned.loc[df_cleaned['Precio_Unitario'] < 0, 'Precio_Unitario'] = np.nan # Reemplazar precios negativos con NaN

# --- Limpieza de la columna 'Cantidad' ---
# Reemplazar valores erróneos (99 y negativos) con NaN
df_cleaned['Cantidad'] = df_cleaned['Cantidad'].replace(99, np.nan)
df_cleaned.loc[df_cleaned['Cantidad'] < 0, 'Cantidad'] = np.nan

# --- Limpieza de la columna 'Estado_Envio' ---
# Estandarizar valores inconsistentes
df_cleaned['Estado_Envio'] = df_cleaned['Estado_Envio'].replace({
    'DESCONOCIDO': 'Pendiente',
    'ERROR_LOGISTICA': 'Cancelado',
    'desconocido': 'Pendiente'
})

# --- Limpieza de la columna 'Region' ---
# Estandarizar 'NA' y 'No Aplica' a un valor consistente o NaN
df_cleaned['Region'] = df_cleaned['Region'].replace(['NA', 'No Aplica'], 'No Definido')

# --- Limpieza de la columna 'Calificacion_Servicio' ---
# Reemplazar valores no numéricos por NaN para permitir el análisis numérico
df_cleaned['Calificacion_Servicio'] = df_cleaned['Calificacion_Servicio'].replace(['MALO', 'ERROR'], np.nan)
df_cleaned['Calificacion_Servicio'] = pd.to_numeric(df_cleaned['Calificacion_Servicio'], errors='coerce')

# --- Limpieza de la columna 'Comentarios_Cliente' ---
# Estandarizar valores que indican falta de información
df_cleaned['Comentarios_Cliente'] = df_cleaned['Comentarios_Cliente'].replace({
    'N/A': 'Sin Comentarios',
    'SIN_INFO': 'Sin Comentarios',
    'Vacío': 'Sin Comentarios',
    'n/a': 'Sin Comentarios'
})


print("\n--- Dataset Después del Reemplazo de Valores (20 Registros) ---\n")
print(df_cleaned)

# 3. En el codigo se debe comprobar que los datos los valores erróneos han sido correctamente reemplazados.
print("\n3. Comprobando que los valores erróneos han sido correctamente reemplazados...\n")

print("--- Conteo de Valores Específicos Después del Reemplazo (Ejemplos) ---\n")
print("Valores en 'Precio_Unitario':\n", df_cleaned['Precio_Unitario'].value_counts(dropna=False))
print("\nValores en 'Cantidad':\n", df_cleaned['Cantidad'].value_counts(dropna=False))
print("\nValores en 'Estado_Envio':\n", df_cleaned['Estado_Envio'].value_counts(dropna=False))
print("\nValores en 'Region':\n", df_cleaned['Region'].value_counts(dropna=False))
print("\nValores en 'Calificacion_Servicio':\n", df_cleaned['Calificacion_Servicio'].value_counts(dropna=False))
print("\nValores en 'Comentarios_Cliente':\n", df_cleaned['Comentarios_Cliente'].value_counts(dropna=False))


# Verificación final: confirmamos la ausencia de los valores "malos" originales
print("\n--- Verificación Detallada de la Ausencia de Valores Erróneos Originales ---\n")

# Para 'Precio_Unitario'
bad_prices = ['NO_APLICA', 'NO_DISP', 'ERROR']
if not df_cleaned['Precio_Unitario'].astype(str).isin(bad_prices).any() and \
   not (df_cleaned['Precio_Unitario'] < 0).any():
    print(f"✔️ Valores {bad_prices} y negativos eliminados/reemplazados de 'Precio_Unitario'.")
else:
    print(f"❌ Aún quedan valores problemáticos en 'Precio_Unitario'.")

# Para 'Cantidad'
if not ((df_cleaned['Cantidad'] == 99) | (df_cleaned['Cantidad'] < 0)).any():
    print("✔️ Valores 99 y negativos eliminados/reemplazados de 'Cantidad'.")
else:
    print("❌ Aún quedan valores 99 o negativos en 'Cantidad'.")

# Para 'Estado_Envio'
bad_statuses = ['DESCONOCIDO', 'ERROR_LOGISTICA', 'desconocido']
if not df_cleaned['Estado_Envio'].isin(bad_statuses).any():
    print(f"✔️ Valores {bad_statuses} eliminados/reemplazados de 'Estado_Envio'.")
else:
    print(f"❌ Aún quedan valores {bad_statuses} en 'Estado_Envio'.")

# Para 'Region'
bad_regions = ['NA', 'No Aplica']
if not df_cleaned['Region'].isin(bad_regions).any():
    print(f"✔️ Valores {bad_regions} eliminados/reemplazados de 'Region'.")
else:
    print(f"❌ Aún quedan valores {bad_regions} en 'Region'.")

# Para 'Calificacion_Servicio'
bad_ratings = ['MALO', 'ERROR']
if not df_cleaned['Calificacion_Servicio'].astype(str).isin(bad_ratings).any() and \
   (df_cleaned['Calificacion_Servicio'].apply(lambda x: isinstance(x, (int, float)))).all(): # Check if all are numeric
    print(f"✔️ Valores {bad_ratings} eliminados/reemplazados de 'Calificacion_Servicio' y convertidos a numérico.")
else:
    print(f"❌ Aún quedan valores problemáticos en 'Calificacion_Servicio'.")

# Para 'Comentarios_Cliente'
bad_comments = ['N/A', 'SIN_INFO', 'Vacío', 'n/a']
if not df_cleaned['Comentarios_Cliente'].isin(bad_comments).any():
    print(f"✔️ Valores {bad_comments} eliminados/reemplazados de 'Comentarios_Cliente'.")
else:
    print(f"❌ Aún quedan valores {bad_comments} en 'Comentarios_Cliente'.")

print("\nProceso de reemplazo de valores completado y verificado.")
```
